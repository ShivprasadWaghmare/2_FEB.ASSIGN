1. Explain with an example each when to use a for loop and a while loop.
ANS:Use a for loop:
when you know in advance how many times you want to repeat a block of code. It's great for iterating over a sequence of elements, like a list.

Example: Printing numbers 1 to 5

for i in range(1, 6):
    print(i)
This loop will run 5 times, printing numbers from 1 to 5.

When to use a while loop:
Use a while loop when you want to repeat a block of code as long as a certain condition is true. It's useful when you don't know in advance how many times the code needs to be repeated.

Example: Printing numbers until a condition is met

counter = 1
while counter <= 5:
    print(counter)
    counter += 1
This loop will run until the counter reaches 6, printing numbers from 1 to 5.

In summary, for loops are for when you know the number of iterations, while loops are for when you're not sure and need to continue until a specific condition is false.

2. Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.
ans:# Sum and Product using For Loop
sum_result = 0
product_result = 1

for i in range(1, 11):
    sum_result += i
    product_result *= i

print("Sum:", sum_result)
print("Product:", product_result)


# Sum and Product using While Loop
sum_result = 0
product_result = 1
count = 1

while count <= 10:
    sum_result += count
    product_result *= count
    count += 1

print("Sum:", sum_result)
print("Product:", product_result)

Create a python program to compute the electricity bill for a household.The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 perunit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user willbe charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.You are required to take the units of electricity consumed in a month from the user as input.Your program must pass this test case: when the unit of electricity consumed by the user in a month is310, the total electricity bill should be 2250.
ans:

def calculate_electricity_bill(units):
    # Define per-unit charges
    charge_1 = 4.5  # for the first 100 units
    charge_2 = 6.0  # for the next 100 units
    charge_3 = 10.0  # for the next 100 units
    charge_4 = 20.0  # for 300 units and above

    # Calculate the electricity bill based on the consumed units
    if units <= 100:
        bill = units * charge_1
    elif units <= 200:
        bill = 100 * charge_1 + (units - 100) * charge_2
    elif units <= 300:
        bill = 100 * charge_1 + 100 * charge_2 + (units - 200) * charge_3
    else:
        bill = 100 * charge_1 + 100 * charge_2 + 100 * charge_3 + (units - 300) * charge_4

    return bill

# Take input from the user
units_consumed = int(input("Enter the units of electricity consumed in a month: "))

# Calculate and display the electricity bill
electricity_bill = calculate_electricity_bill(units_consumed)
print(f"The total electricity bill is: Rs. {electricity_bill}")

4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of eachnumber and if the cube of that number is divisible by 4 or 5 then append that number in a list and printthat list.
ans:
# Using a for loop
cubes_for = []
for num in range(1, 101):
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_for.append(num)

print("Using for loop:", cubes_for)

# Using a while loop
cubes_while = []
num = 1
while num <= 100:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_while.append(num)
    num += 1

print("Using while loop:", cubes_while)
